```python
# ch 2 - 연습문제 7

import turtle #터틀 그래픽 모듈 

 

t = turtle.Turtle() #거북이 생성

t.shape("turtle") #거북이 모양 결정

 

side = 50 # side 변수를 50로



#상단 우측 사각형

t.forward(side) #거북이가 앞으로 50만큼 움직인다.

t.left(90) #거북이가 왼쪽으로 90도 회전한다.

t.forward(side)#거북이가 앞으로 50만큼 움직인다.

t.left(90) #거북이가 왼쪽으로 90도 회전한다.

t.forward(side)#거북이가 앞으로 50만큼 움직인다.

t.left(90) #거북이가 왼쪽으로 90도 회전한다.

t.forward(side) #거북이가 앞으로 50만큼 움직인다.

 

#하단 우측 사각형

t.forward(side) #거북이가 앞으로 50만큼 움직인다.

t.left(90) #거북이가 왼쪽으로 90도 회전한다.

t.forward(side) #거북이가 앞으로 50만큼 움직인다.

t.left(90) #거북이가 왼쪽으로 90도 회전한다.

t.forward(side) #거북이가 앞으로 50만큼 움직인다.

t.left(90) #거북이가 왼쪽으로 90도 회전한다.

 

#상단 좌측 사각형

t.forward(side) #거북이가 앞으로 50만큼 움직인다.

t.forward(side) #거북이가 앞으로 50만큼 움직인다.

t.right(90) #거북이가 오른쪽으로 90도 회전한다.

t.forward(side) #거북이가 앞으로 50만큼 움직인다.

t.right(90) #거북이가 오른쪽으로 90도 회전한다.

t.forward(side) #거북이가 앞으로 50만큼 움직인다.

t.right(90) #거북이가 오른쪽으로 90도 회전한다.

 

#하단 좌측 사각형

t.forward(side) #거북이가 앞으로 50만큼 움직인다.

t.forward(side) #거북이가 앞으로 50만큼 움직인다.

t.right(90) #거북이가 오른쪽으로 90도 회전한다.

t.forward(side) #거북이가 앞으로 50만큼 움직인다.

t.right(90) #거북이가 오른쪽으로 90도 회전한다.

t.forward(side) #거북이가 앞으로 50만큼 움직인다.

```

```python
# ch 4 - 연습문제 7 

color_list=[] #공백 list를 생성한다.

color=input("색상 #1을 입력하시오: ") #색상 1을 입력받는다.

color_list.append(color) #append 함수를 통해 입력받은 색상을 list에 추가한다.

color=input("색상 #2을 입력하시오: ")#색상 2를 입력받는다.

color_list.append(color)#append 함수를 통해 입력받은 색상을 list에 추가한다.

color=input("색상 #3을 입력하시오: ")#색상 3을 입력받는다.

color_list.append(color)#append 함수를 통해 입력받은 색상을 list에 추가한다.

 

import turtle  #이것을 통해 터틀 그래픽 모듈을 불러들임.

 

t = turtle.Turtle() #거북이를 생성한다.

t.shape("turtle") #거북이 모양을 정한다. 필수X. 다른 모양도 OK.

 

t.fillcolor(color_list[0]) # 입력받은 색상1을 채우기 색으로 지정한다.

t.begin_fill() #채우기 시작한다.

t.circle(50) #반지름이 50픽셀인 원을 그림.

t.end_fill() #채우기를 끝냄. 

t.up() #펜을 들고있는 상태. 즉, 움직여도 그려지지않는다.

 

t.goto(100, 0) #x 좌표가 100이고 y좌표가 0인 지점에서 그리기를 시작한다.

t.down() #펜을 내려놓은 상태. 즉, 움직이면 그려진다.

t.fillcolor(color_list[1])

t.begin_fill() #채우기 시작한다.

t.circle(50) #반지름이 50픽셀인 원을 그린다.

t.end_fill() #채우기를 끝낸다. 

t.up() #펜을 들고있는 상태. 즉, 움직여도 그려지지않는다.

 

t.goto(200, 0) #x좌표가 200이고 y좌표가 0인 지점에서 그리기를 시작한다.

t.down() #펜을 내려놓은 상태여서 움직이면 그려진다.

t.fillcolor(color_list[2])

t.begin_fill() #채우기 시작한다.

t.circle(50) #반지름이 50픽셀인 원을 그린다.

t.end_fill() #채우기를 끝낸다. 
```

```python
# ch 5 - 연습문제 7

import random #random 모듈을 불러들인다.

ans = random.randint(10, 99) #10부터 99까지의 난수를 생성하는 함수이다. 

num = int(input("복권번호를 입력하시요(10에서 99 사이) : ")) #10부터 99사이의 숫자를 입력받는다. 

 

 

print("당첨번호는" + str(ans) + "입니다.") #ans에 저장된 난수를 출력한다.

if num < 10 : #만약 입력받은 숫자가 10보다 작으면 

    print("한자리 수는 복권번호로 선택하실 수 없습니다.") #문장을 출력한다.

elif ans == num : #만약 입력받은 숫자가 생성된 난수와 같으면 

    print("상금은 100만원 입니다.") #문장을 출력한다.

elif ((num//10 == ans//10) or (num//10 == ans %10)) : #입력값을 10으로 나눈 몫과 난수를 10으로 나눈 몫이 같거나 또는, 입력값을 10으로 나눈 몫과 난수를 10으로 나눈 나머지가 같다면 

    print("상금은 50만원 입니다.") #문장을 출력한다.

elif ((num%10 == ans//10) or (num%10 == ans %10)) :#입력값을 10으로 나눈 나머지와 난수를 10으로 나눈 몫이 같거나 또는, 입력값을 10으로 나눈 나머지와 난수를 10으로 나눈 나머지와 같다면 

    print("상금은 50만원 입니다.") #이 문장을 출력한다.

else :

    print("당첨되지 않았습니다.") #이 문장을 출력한다.
```
