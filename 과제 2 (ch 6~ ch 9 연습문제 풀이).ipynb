```python
# ch6-연습문제 7
# 눈모양

# import를 하는 이유-파이썬에 기본적으로 없는 함수 사용을 위해

import turtle # 거북이 모듈로 지정
t = turtle.Turtle() 
t.shape("turtle") # 그래픽 모양을 거북이로 지정
t.left(90) # 방향을 왼쪽으로 90 만큼 회전한다
t.color("blue") #선 색을 파란색으로 한다. 
                # 참고로 t.bgcolor() 는 배경색을 나타낸다.

for i in range(6) : # 하나의 눈 모양을 그리기 위해 반복문 실행. 행동 i 에 대한 범위는 0~5 로, 총6번을 실행한다.
    
    t.forward(100) # 100 픽셀 직진
    t.forward(-30) # 30 픽셀 후진(-)
    t.left(60) # 방향을 왼쪽으로 60도 회전
    t.forward(30) 
    t.forward(-30) 
    t.right(120) # 방향을 오른쪽으로 120도 회전
    t.forward(30) 
    t.forward(-30) 
    
    # 하나의 눈 모양을 그린 후, 중간으로 돌아가기.
    t.left(60) 
    t.forward(-70) #70픽셀 만큼 후진. forward(-70) 과 같은 기능.
    
    #다음 눈꽃 모양을 그리기 위한 각도 조정을 한다.
    t.right(60) 

    
```

```python
# ch7-연습문제 7
# 원의 면적과 둘레


# 먼저 지정된 반지름 값의 원에 대한 면적, 둘레 값 출력됨.

def circleArea(radius): # 키워드 def 를 사용해 circleArea 함수 정의 
    result = PI*radius**2   # 결과값에 pi*반지름^2  을 반환한다.
    
    print("반지름이 " + str(radius) + "인 원의 면적 :" + str(result)) 

def circleCircumference(radius): #circleCircumference 함수 정의.
     result = 2*PI*radius  # 결과값에 2*pi*반지름 값 반환.
        
     print("반지름이 " + str(radius) + "인 원의 둘레 :" + str(result))


PI = 3.14 # pi를 3.14로 초기화.
circleArea(5) # circle Area 함수의 파라미터에 5 지정 --> 반지름 5 에 대해 면적 함수 실행  
circleCircumference(5) # circleCircumference 함수의 파라미터에 5 지정 --> 반지름 5 에 대해 둘레 함수 실행

 
# 다음으로, 사용자가 입력한 반지름 값의 원에 대한 면적, 둘레 값 출력됨.  


def circleArea(radius):
    result = PI*radius**2
    print("반지름이 " + str(radius) + "인 원의 면적 :" + str(result))

def circleCircumference(radius):
     result = 2*PI*radius
     print("반지름이 " + str(radius) + "인 원의 둘레 :" + str(result))

r = float(input("반지름 : ")) # 반지름 값을 입력받음.
PI = 3.14
circleArea(r) # 입력받은 반지름 값에 대한 면적 함수 실행
circleCircumference(r)  # 입력받은 반지름 값에 대한 둘레 함수 실행


```

```python
# ch9-연습문제 2
# 주사위 빈도 계산

import random # 랜덤 모듈 import

counters= [0,0,0,0,0,0] # 주사위 값이 나오는 빈도를 리스트에 저장하기 위해 6칸의 빈 리스트 생성.

for i in range(1000): # 행동 i 에 대한 범위 0~999 인 반복문 실행
    
    ran = random.randint(1,6) #1이상 6 이하의 정수를 랜덤하게 생성.
    if ran == 1:
        counters[0] = counters[0] + 1  # 리스트 0번째 값을 1 증가.
    elif ran == 2:
        counters[1] = counters[1] + 1  # 리스트 1번째 값 1 증가.
    elif ran == 3:
        counters[2] = counters[2] + 1  # 리스트 2번째 값 1 증가
    elif ran == 4:
        counters[3] = counters[3] + 1  # 리스트 3번째 값 1 증가
    elif ran == 5:
        counters[4] = counters[4] + 1  # 리스트 4번째 값 1 증가
    elif ran == 6:
        counters[5] = counters[5] + 1  # 리스트 5번째 값 1 증가

for j in range(6): # 행동 j 에 대한 범위 0~5 인 반복문.
    print("주사위가 %s 인 경우는  %s " % (j+1, counters[j])) # 다음 문장 출력. (j 값을 활용.)
```

```python

# ch9-연습문제 7
# 도메인

domains = {"kr" : "대한민국", "sk" : "슬로바키아", "no" : "노르웨이", "us" : "미국", "jp" : "일본", "hu" : "헝가리", "de" : "독일"}

# 인터넷 도메인의 약자와 해당하는 국가를 딕셔너리에 저장.

for k, v in domains.items(): # k,v에 대한 반복분 실행.
    print(k, ":", v) # 딕셔너리에서 k(키=약자)와 v (값=국가) 를 꺼내어 출력.
```
